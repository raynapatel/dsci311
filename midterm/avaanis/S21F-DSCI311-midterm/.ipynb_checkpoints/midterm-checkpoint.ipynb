### I've provided a file named WebLogFormat.pdf which describes the format of these entries.

Now define a regular expression named `ip_pat` that will match the ipv4 address at the beginning of a log entry, assign to a variable named `ipaddr` the result of applying `ip_pat` to `first`

```python
# 2 points for this, one for correct regular expression, one for successfully extracting the ip address from first
ip_pat = r'((?<!\s)\d\d.\d+\.\d+.\d+)'
ipaddr = re.search(ip_pat, first).group()
ipaddr
```

I have also provided you with a CSV file named `addrs.csv` which maps from IP addresses to hostname and latitude/longitude coordinates for that IP address. Create a dataframe named `addr_map` from `addrs.csv`.

```python
# 1 point for this
addr_map = pd.read_csv('addrs.csv')
addr_map
```

Now locate the row in `addr_map` that matches `ipaddr`, convert it to a Python list, and assign this list to a variable named `row`.

```python
# 1 point for this
ipaddrbool = addr_map['ip'] == ipaddr
row = addr_map[ipaddrbool]
row = row.values.tolist()
row = row[0]
row
```

Now extract the timestamp from `first` yielding three strings: the date in the format `dd/Mmm/yyyy`, the time in the format `hh:mm:ss`, and the timezone in the format `-|+hhmm`; assign these three strings to variables named `date`, `time`, and `timezone`.

```python
# 3 points for this, one each for the date, time, and timezone
pattern = r'\[(\d+\/\w+\/\d+):(\d+:\d+:\d+).(-\d+)]'
date, time, timezone = re.search(pattern, first).groups()
date, time, timezone
```

Now define a function named `log_entry_to_series` that has one argument, a log entry line, and yields a pandas Series that is indexed by 'ip', 'host', 'latitude', 'longitude', 'date', 'time', and 'timezone'

```python
# 5 points for this; it's just an assemblage of everything you've done above
def log_entry_to_series(line):
    ip_pat = r'((?<!\s)\d+.\d+\.\d+.\d+)'
    ipaddr = re.search(ip_pat, line).group()
    pattern = r'\[(\d+\/\w+\/\d+):(\d+:\d+:\d+).(-\d+)]'
    date, time, timezone = re.search(pattern, line).groups()   
    ipaddrbool = addr_map['ip'] == ipaddr
    row = addr_map[ipaddrbool]
    row = row.values.tolist()
    if len(row) == 0:
        host = 'NaN'
        lat = 'NaN'
        long = 'NaN'
    else:
        row = row[0]
        host = row[1]
        lat = row[2]
        long = row[3]
    d = {'ip': ipaddr, 'host': host, 'latitude': lat, 'longitude': long, 'date': date, 'time': time, 'timezone':timezone}
    return pd.Series(data = d, index = ['ip', 'host', 'latitude', 'longitude', 'date', 'time', 'timezone'])

log_entry_to_series(first)
```

Now create a dataframe named `log_info` with columns 'ip', 'host', 'latitude', 'longitude', 'date', 'time', and 'timezone'

```python
cols = ['ip', 'host', 'latitude', 'longitude', 'date', 'time', 'timezone']

log_info = pd.DataFrame(columns=cols)
for i in (np.arange(len(log_lines)-1)):
    to_append = log_entry_to_series(log_lines[i])
    to_append.to_frame()
    log_info = log_info.append(to_append, ignore_index = True)

log_info
```

Now we can begin to analyze the data in `log_info`. First we'd like to know the ip addresses that are most represented in the dataframe. Create a dataframe named `top` consisting of two columns, 'ip' and 'count', that has the rows sorted in descending order by number of occurrences.

```python
# 2 points for this, one for getting the column names right, the other for having the rows sorted in descending order
ip_select = pd.DataFrame(log_info['ip'])
top = ip_select.value_counts(ip_select['ip']).to_frame(name = 'count')
top = top.reset_index()
top
```

Now let's figure out which was the busiest day for the web server. Do a similar kind of analysis to that above to yield a dataframe named `top_day` which contains two columns ('date' and 'count') that is sorted by descending order of count.

```python
# 2 points for this, again 1 for correct column labels, 1 for dataframe sorted in descending order of count.
date_select = pd.DataFrame(log_info['date'])
top_day = date_select.value_counts(date_select['date']).to_frame(name = 'count')

top_day = top_day.reset_index()

top_day
```

Which is the busiest hour of the day for the web server? Create a new dataframe named `top_hour` that consists of two columns named 'hour' and 'count', with the dataframe sorted in descending order by count.

```python
# 3 points for this one: 1 for correct column labels, 1 for dataframe sorted correctly, and 1 for the intermediate
# computations you will need to do

hour_select = pd.DataFrame(log_info['time'])
hour_select['hour'] = (hour_select['time']
                      .str.replace('\:\d\d\:\d\d', " ")
                      .str.strip())
top_hour = hour_select.value_counts(hour_select['hour']).to_frame(name = 'count')
top_hour = top_hour.reset_index()

top_hour.head()
```
